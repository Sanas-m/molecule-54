```python
import pandas as pd
from numpy import float64, subtract, divide
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def convert_rep(row):
    row['rep'] = row['Sample Name'][-1]
    row['Sample Name'] = row['Sample Name'][:len(row['Sample Name']) - 5]
    return row
    
```

```python
rawDataSet_3and4 = pd.read_excel(r'/Users/sanas/Projects/molecule-54/AngII-stim/resources/AngII_rep3-rep4.xls', sheet_name = 'Results', header = 50, decimal = ",")

AngII_clean_3and4 = rawDataSet_3and4.drop(columns = ['Well', 'Well Position', 'Omit', 'Task', 'Reporter', 'Quencher', 'Quantity', \
                                    'Quantity Mean', 'Quantity SD', 'RQ', 'RQ Min', 'RQ Max', 'Delta Ct', 'Delta Ct Mean', \
                                     'Delta Ct SD', 'Delta Ct SE', 'Delta Delta Ct', 'Automatic Ct Threshold', \
                                     'Ct Threshold', 'Automatic Baseline', 'Baseline Start', \
                                    'Baseline End', 'Amp Status', 'Comments', 'Cq Conf', 'CQCONF', 'EXPFAIL', 'NOAMP',])

AngII_clean_3and4 = AngII_clean_3and4.dropna(axis = 0)
AngII_clean_3and4 = AngII_clean_3and4[AngII_clean_3and4['Sample Name'].str.contains('rep')]
AngII_clean_3and4 = AngII_clean_3and4.apply(convert_rep, axis = 1)

AngII_clean_3and4
```

```python
rawDataSet_5and6 = pd.read_excel(r'/Users/sanas/Projects/molecule-54/AngII-stim/resources/AngII_rep5-rep6.xls', sheet_name = 'Results', header = 50, decimal = ",")

AngII_clean_5and6 = rawDataSet_5and6.drop(columns = ['Well', 'Well Position', 'Omit', 'Task', 'Reporter', 'Quencher', 'Quantity', \
                                    'Quantity Mean', 'Quantity SD', 'RQ', 'RQ Min', 'RQ Max', 'Delta Ct', 'Delta Ct Mean', \
                                     'Delta Ct SD', 'Delta Ct SE', 'Delta Delta Ct', 'Automatic Ct Threshold', \
                                     'Ct Threshold', 'Automatic Baseline', 'Baseline Start', \
                                    'Baseline End', 'Amp Status', 'Comments', 'Cq Conf', 'CQCONF', 'EXPFAIL']) 

AngII_clean_5and6 = AngII_clean_5and6.dropna(axis = 0)
AngII_clean_5and6 = AngII_clean_5and6[AngII_clean_5and6['Sample Name'].str.contains('rep')]
AngII_clean_5and6 = AngII_clean_5and6.apply(convert_rep, axis = 1)

AngII_clean_5and6
```

```python
mergedAngII = pd.concat([AngII_wo_na_2, AngII_clean_3and4, AngII_clean_5and6]).replace('Untransfected', 'Untreated', regex = True).replace('mol-54', 'm-54', regex = True)
mergedAngII
```

```python
AngII_data_dedupl = mergedAngII.drop(columns=['CT', 'Ct SD']).drop_duplicates()
AngII_data_dedupl
```

```python
CYP11B2_values = AngII_data_dedupl[AngII_data_dedupl['Target Name'] == 'CYP11B2']

def delta_CT(CYP11B2_value):
    GAPDH_value = AngII_data_dedupl[(AngII_data_dedupl['Target Name'] == 'GAPDH') &
                                    (AngII_data_dedupl['Sample Name'] == CYP11B2_value['Sample Name']) &
                                    (AngII_data_dedupl['rep'] == CYP11B2_value['rep'])
                                  ]
    CYP11B2_value['GAPDH'] = float64(GAPDH_value['Ct Mean'])
    CYP11B2_value['Delta CT'] = float64(subtract(CYP11B2_value['Ct Mean'], GAPDH_value['Ct Mean']))   
    
    return CYP11B2_value

deltaCT_CYP11B2 = CYP11B2_values.apply(delta_CT, axis = 1).drop(columns = ['Target Name', 'GAPDH'])
deltaCT_CYP11B2
```

```python
doubleDeltaCT_CYP11B2 = 2 ** (-deltaCT_CYP11B2['Delta CT'])
float64(doubleDeltaCT_CYP11B2)
```

```python
deltaCT_CYP11B2['Double Delta CT'] = doubleDeltaCT_CYP11B2
deltaCT_CYP11B2
```

### Manual division

```python
def extract_untreated (browsed_dataframe):
    untreated = browsed_dataframe[(browsed_dataframe['Sample Name'] == 'Untreated') &
                          (browsed_dataframe['rep'] == browsed_dataframe['rep'])
                         ]
    return untreated

untreated_per_rep = extract_untreated(deltaCT_CYP11B2).set_index('rep')
untreated_per_rep
```

```python
def extract_angII (browsed_dataframe):
    angII = browsed_dataframe[(browsed_dataframe['Sample Name'] == 'AngII') &
                          (browsed_dataframe['rep'] == browsed_dataframe['rep'])
                         ]
    return angII

angII_per_rep = extract_angII(deltaCT_CYP11B2).set_index('rep')
angII_per_rep
```

```python
def extract_angII_m54 (browsed_dataframe):
    angII_m54 = browsed_dataframe[(browsed_dataframe['Sample Name'] == 'AngII + m-54') &
                          (browsed_dataframe['rep'] == browsed_dataframe['rep'])
                         ]
    return angII_m54

angII_m54_per_rep = extract_angII_m54(deltaCT_CYP11B2).set_index('rep')
angII_m54_per_rep
```

```python
angII_per_rep['Untreated double delta'] = untreated_per_rep['Double Delta CT']
angII_per_rep
```

```python
angII_per_rep['result'] = angII_per_rep['Double Delta CT'].div(angII_per_rep['Untreated double delta'].values,axis=0)
angII_per_rep
```

```python
untreated_rep2 = deltaCT_CYP11B2[deltaCT_CYP11B2['rep'] == '2']

untreated_rep3 = deltaCT_CYP11B2[deltaCT_CYP11B2['rep'] == '3']

untreated_rep4 = deltaCT_CYP11B2[deltaCT_CYP11B2['rep'] == '4']

untreated_rep5 = deltaCT_CYP11B2[deltaCT_CYP11B2['rep'] == '5']

untreated_rep6 = deltaCT_CYP11B2[deltaCT_CYP11B2['rep'] == '6']

def foldChange(rep_data_frame):
    untreated = rep_data_frame[rep_data_frame['Sample Name'] == 'Untreated']['Double Delta CT']
    angII = rep_data_frame[rep_data_frame['Sample Name'] == 'AngII']['Double Delta CT']
    angII_m54 = rep_data_frame[rep_data_frame['Sample Name'] == 'AngII + m-54']['Double Delta CT']

    foldChange_angII = divide(float64(angII), float64(untreated))
    foldChange_angIIm54 = divide(float64(angII_m54), float64(untreated))
            
    return [1, foldChange_angII, foldChange_angIIm54]
    
untreated_rep2['Fold Change'] = foldChange(untreated_rep2)
untreated_rep3['Fold Change'] = foldChange(untreated_rep3)
untreated_rep4['Fold Change'] = foldChange(untreated_rep4)
untreated_rep5['Fold Change'] = foldChange(untreated_rep5)
untreated_rep6['Fold Change'] = foldChange(untreated_rep6)
combined = pd.concat([untreated_rep2, untreated_rep3, untreated_rep4, untreated_rep5, untreated_rep6])
combined
```

```python
mean = combined.groupby(['Sample Name'])['Fold Change'].mean()
plot_mean = mean.to_frame().rename(columns = {"Fold Change": "Mean Fold Change"})
plot_mean
```

```python
std = combined.groupby(['Sample Name'])['Fold Change'].std()
plot_std = std.to_frame().rename(columns = {"Fold Change": "Std Fold Change"})
plot_std
```

```python
plot_data = pd.concat([plot_mean, plot_std], axis = 1)
plot_data
```

```python
plot_data.plot(kind = 'bar', legend = False, title = 'AngII', yerr = 'Std Fold Change')
```
