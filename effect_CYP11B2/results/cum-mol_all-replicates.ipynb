# Effect of Molecule-54 On CYP11B2 Expression

```python
import pandas as pd
from numpy import float64, subtract
```

```python
def convert_rep(row):
    row['rep'] = row['Sample Name'][-1]
    row['Sample Name'] = row['Sample Name'][:len(row['Sample Name']) - 5]
    return row
```

```python
rawDataSet_1 = pd.read_excel(r'/Users/sanas/Projects/molecule-54/effect_CYP11B2/resources/beta_R1.xls', sheet_name = 'Results', header = 50, decimal = ",")

cum_mol_clean_1 = rawDataSet_1.drop(columns = ['Well', 'Well Position', 'Omit', 'Task', 'Reporter', 'Quencher', 'Quantity', \
                                    'Quantity Mean', 'Quantity SD', 'RQ', 'RQ Min', 'RQ Max', 'Delta Ct', 'Delta Ct Mean', \
                                     'Delta Ct SD', 'Delta Ct SE', 'Delta Delta Ct', 'Automatic Ct Threshold', \
                                     'Ct Threshold', 'Automatic Baseline', 'Baseline Start', \
                                    'Baseline End', 'Amp Status', 'Comments', 'Cq Conf', 'CQCONF', 'EXPFAIL'])

cum_mol_clean_1['rep'] = '1'
cum_mol_clean_1 = cum_mol_clean_1[cum_mol_clean_1['Sample Name'].str.contains('rep')== False ]
cum_mol_clean_1 = cum_mol_clean_1.dropna(axis = 0)
cum_mol_clean_1
```

```python
rawDataSet_2 = pd.read_excel(r'/Users/sanas/Projects/molecule-54/effect_CYP11B2/resources/beta_R2-R3.xls', sheet_name = 'Results', header = 50, decimal = ",")

cum_mol_clean_2 = rawDataSet_2.drop(columns = ['Well', 'Well Position', 'Omit', 'Task', 'Reporter', 'Quencher', 'Quantity', \
                                    'Quantity Mean', 'Quantity SD', 'RQ', 'RQ Min', 'RQ Max', 'Delta Ct', 'Delta Ct Mean', \
                                     'Delta Ct SD', 'Delta Ct SE', 'Delta Delta Ct', 'Automatic Ct Threshold', \
                                     'Ct Threshold', 'Automatic Baseline', 'Baseline Start', \
                                    'Baseline End', 'Amp Status', 'Comments', 'Cq Conf', 'CQCONF', 'EXPFAIL', 'HIGHSD'])


cum_mol_clean_2 = cum_mol_clean_2[cum_mol_clean_2['Sample Name'].str.contains('rep')== True ]
cum_mol_clean_2 = cum_mol_clean_2.apply(convert_rep, axis = 1)
cum_mol_clean_2 = cum_mol_clean_2.dropna(axis = 0)
cum_mol_clean_2

#ask Severin why 2 and 3 is automatically added
```

```python
merged_cum_mol = pd.concat([cum_mol_clean_1, cum_mol_clean_2])
merged_cum_mol
```

```python
cum_mol_dedupl = merged_cum_mol.drop(columns=['CT', 'Ct SD']).drop_duplicates().replace('\+ m-54', '+m-54', regex = True).replace('\- m-54', '-m-54', regex = True).replace('\+ cum', '+cum', regex = True).replace('\- cum', '-cum', regex = True)

cum_mol_dedupl
```

```python
# defining control = calibrator samples for GAPDH per replicate

rep_1 = '1'
rep_2 = '2'
rep_3 = '3'

target_GAPDH = 'GAPDH'
target_KCNJ5 = 'KCNJ5'
target_CYP11B2 = 'CYP11B2'

def control (replicate, target):
    ctrl = cum_mol_dedupl[(cum_mol_dedupl['Sample Name'] == '-cum +m-54') &
                          (cum_mol_dedupl['Target Name'] == target) &
                          (cum_mol_dedupl['rep'] == replicate)].reset_index().iloc[0, 3]
    return ctrl


ctrl_GAPDH_rep1 = control(rep_1, target_GAPDH)
ctrl_GAPDH_rep2 = control(rep_2, target_GAPDH)
ctrl_GAPDH_rep3 = control(rep_3, target_GAPDH)

ctrl_KCNJ5_rep1 = control(rep_1, target_KCNJ5)
ctrl_KCNJ5_rep2 = control(rep_2, target_KCNJ5)
ctrl_KCNJ5_rep3 = control(rep_3, target_KCNJ5)

ctrl_CYP11B2_rep1 = control(rep_1, target_CYP11B2)
ctrl_CYP11B2_rep2 = control(rep_2, target_CYP11B2)
ctrl_CYP11B2_rep3 = control(rep_3, target_CYP11B2)
```

```python
# Extracting dataframes per replicate for each probe

def dataframes (replicate, target):
    dataframe = cum_mol_dedupl[(cum_mol_dedupl['Target Name'] == target) &
                              (cum_mol_dedupl['rep'] == replicate)]
    return dataframe


GAPDH_rep1 = dataframes(rep_1, target_GAPDH)['Ct Mean']
GAPDH_rep2 = dataframes(rep_2, target_GAPDH)
GAPDH_rep3 = dataframes(rep_3, target_GAPDH)

KCNJ5_rep1 = dataframes(rep_1, target_KCNJ5)['Ct Mean']
KCNJ5_rep2 = dataframes(rep_2, target_KCNJ5)
KCNJ5_rep3 = dataframes(rep_3, target_KCNJ5)

CYP11B2_rep1 = dataframes(rep_1, target_CYP11B2)['Ct Mean']
CYP11B2_rep2 = dataframes(rep_2, target_CYP11B2)
CYP11B2_rep3 = dataframes(rep_3, target_CYP11B2)

```

```python
CYP11B2_values = cum_mol_dedupl[cum_mol_dedupl['Target Name'] == 'CYP11B2']

def delta_CT(CYP11B2):
    GAPDH = cum_mol_dedupl[(cum_mol_dedupl['Target Name'] == 'GAPDH') &
                            (cum_mol_dedupl['Sample Name'] == CYP11B2['Sample Name']) &
                            (cum_mol_dedupl['rep'] == CYP11B2['rep'])
                          ]
    CYP11B2['GAPDH'] = float64(GAPDH['Ct Mean'])
    CYP11B2['dCt CYP11B2'] = float64(subtract(CYP11B2['Ct Mean'], GAPDH['Ct Mean']))   
    
    return CYP11B2

deltaCT_CYP11B2 = CYP11B2_values.apply(delta_CT, axis = 1).drop(columns = ['Target Name', 'GAPDH'])
clean_deltaCT_CYP11B2 = deltaCT_CYP11B2.reset_index().drop(columns=['index'])
clean_deltaCT_CYP11B2
```

```python
KCNJ5_values = cum_mol_dedupl[cum_mol_dedupl['Target Name'] == 'KCNJ5']

def delta_CT(KCNJ5):
    GAPDH = cum_mol_dedupl[(cum_mol_dedupl['Target Name'] == 'GAPDH') &
                            (cum_mol_dedupl['Sample Name'] == KCNJ5['Sample Name']) &
                            (cum_mol_dedupl['rep'] == KCNJ5['rep'])
                          ]
    KCNJ5['GAPDH'] = float64(GAPDH['Ct Mean'])
    KCNJ5['dCt KCNJ5'] = float64(subtract(KCNJ5['Ct Mean'], GAPDH['Ct Mean']))   
    
    return KCNJ5

deltaCT_KCNJ5 = KCNJ5_values.apply(delta_CT, axis = 1).drop(columns = ['Target Name', 'GAPDH'])
clean_deltaCT_KCNJ5 = deltaCT_KCNJ5.reset_index().drop(columns=['index'])
clean_deltaCT_KCNJ5
```

```python
# defining control

def control (dataframe, replicate):
    ctrl = dataframe[(dataframe['Sample Name'] == '-cum +m-54') &
                (dataframe['rep'] == replicate)]
    return ctrl


ctrl_KCNJ5_rep1 = control(clean_deltaCT_KCNJ5, '1')['dCt KCNJ5']
ctrl_KCNJ5_rep2 = control(clean_deltaCT_KCNJ5, '2')['dCt KCNJ5']
ctrl_KCNJ5_rep3 = control(clean_deltaCT_KCNJ5, '3')['dCt KCNJ5']

ctrl_CYP11B2_rep1 = control(clean_deltaCT_CYP11B2, '1')['dCt CYP11B2']
ctrl_CYP11B2_rep2 = control(clean_deltaCT_CYP11B2, '2')['dCt CYP11B2']
ctrl_CYP11B2_rep3 = control(clean_deltaCT_CYP11B2, '3')['dCt CYP11B2']
```

```python
clean_deltaCT_KCNJ5
```

```python
# Subtract KCNJ5/CYP11B2 values with GAPDH value

results_KCNJ5 = []
results_CYP11B2 = []

def subtract_ctrl (dataframe_defined, ctrl_defined, start, end, results_defined):
    dataframe_looped = dataframe_defined.iloc[start:end]
    ctrl_looped = ctrl_defined
    results = results_defined
    for i in dataframe_looped:
        ddCt = i - ctrl_looped
        results.append(ddCt)
    return results

ddCt_KCNJ5 = subtract_ctrl(clean_deltaCT_KCNJ5['dCt KCNJ5'], ctrl_KCNJ5_rep1, 0, 3, results_KCNJ5)
ddCt_KCNJ5 = subtract_ctrl(clean_deltaCT_KCNJ5['dCt KCNJ5'], ctrl_KCNJ5_rep2, 3, 6, results_KCNJ5)
ddCt_KCNJ5 = subtract_ctrl(clean_deltaCT_KCNJ5['dCt KCNJ5'], ctrl_KCNJ5_rep3, 6, 9, results_KCNJ5)

ddCt_CYP11B2 = subtract_ctrl(clean_deltaCT_CYP11B2['dCt CYP11B2'], ctrl_CYP11B2_rep1, 0, 3, results_CYP11B2)
ddCt_CYP11B2 = subtract_ctrl(clean_deltaCT_CYP11B2['dCt CYP11B2'], ctrl_CYP11B2_rep2, 3, 6, results_CYP11B2)
ddCt_CYP11B2 = subtract_ctrl(clean_deltaCT_CYP11B2['dCt CYP11B2'], ctrl_CYP11B2_rep3, 6, 9, results_CYP11B2)
ddCt_CYP11B2
```

```python
clean_deltaCT_CYP11B2['dCt KCNJ5'] = clean_deltaCT_KCNJ5['dCt KCNJ5']
all_results = clean_deltaCT_CYP11B2.drop(columns = ['Ct Mean'])
all_results['ddCt CYP11B2'] = float64(ddCt_CYP11B2)
all_results['ddCt KCNJ5'] = float64(ddCt_KCNJ5)
all_results
```

```python
all_results['Fold Change CYP11B2'] = 2**(-all_results['ddCt CYP11B2'])
all_results['Fold Change KCNJ5'] = 2**(-all_results['ddCt KCNJ5'])
all_results
```
